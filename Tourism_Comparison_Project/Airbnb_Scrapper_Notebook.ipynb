## Combine all scrapped airbnb data in each city into 1 dataframe and export to csv file

```python
import os
import pandas as pd

create dataframes of airbnb listing for each city
beirut_listing = pd.DataFrame()
barcelona_listing = pd.DataFrame()
dubrovnik_listing = pd.DataFrame()
amsterdam_listing = pd.DataFrame()
dubai_listing = pd.DataFrame()
cyprus_listing = pd.DataFrame()
lisbon_listing = pd.DataFrame()

# loop into the 40 scrapped listing files for each city and append them to their corresponding city dataframe 
for i in range(41): #40 excel files scraped
    file_beirut = 'beirut_'+str(i)+'.xlsx'
    file_barcelona = 'barcelona_'+str(i)+'.xlsx'
    file_dubrovnik = 'dubrovnik_'+str(i)+'.xlsx'
    file_amsterdam = 'amsterdam_'+str(i)+'.xlsx'
    file_dubai= 'dubai_'+str(i)+'.xlsx'
    file_cyprus = 'cyprus_'+str(i)+'.xlsx'
    file_lisbon = 'lisbon_'+str(i)+'.xlsx'

    # get size of each file 
    size_beirut = os.path.getsize(file_beirut)
    size_barcelona = os.path.getsize(file_barcelona)
    size_dubrovnik = os.path.getsize(file_dubrovnik)
    size_amsterdam = os.path.getsize(file_amsterdam)
    size_dubai = os.path.getsize(file_dubai)
    size_cyprus = os.path.getsize(file_cyprus)
    size_lisbon = os.path.getsize(file_lisbon)
    
    #check if file has no entry then append to dataframe
    if size_beirut != 0:
        beirut_listing = beirut_listing.append(pd.read_excel(file_beirut))
    if size_barcelona != 0:
        barcelona_listing = barcelona_listing.append(pd.read_excel(file_barcelona))
    if size_dubrovnik != 0:
        dubrovnik_listing = dubrovnik_listing.append(pd.read_excel(file_dubrovnik))
    if size_amsterdam != 0:
        amsterdam_listing = amsterdam_listing.append(pd.read_excel(file_amsterdam))
    if size_dubai != 0:
        dubai_listing = dubai_listing.append(pd.read_excel(file_dubai))
    if size_cyprus != 0:
        cyprus_listing = cyprus_listing.append(pd.read_excel(file_cyprus))
    if size_lisbon != 0:
        lisbon_listing = lisbon_listing.append(pd.read_excel(file_lisbon))
     
    
beirut_listing.to_csv('beirut_listing_july_2020.csv')
barcelona_listing.to_csv('barcelona_listing_july_2020.csv')
dubrovnik_listing.to_csv('dubrovnik_listing_july_2020.csv')
amsterdam_listing.to_csv('amsterdam_listing_july_2020.csv')
dubai_listing.to_csv('duabi_listing_july_2020.csv')
cyprus_listing.to_csv('cyprus_listing_july_2020.csv')
lisbon_listing.to_csv('lisbon_listing_july_2020.csv')
```
